Predicting Final Result wrt to the Selected Attributes from the Dataset Using RandomForest Classifier Model and we will also predict other use cases like age group who wants to access course.


```python
import pandas as pd
import numpy as np

import tensorflow as tf

import tensorflow.keras as keras

from seaborn import heatmap

from keras import Input
from keras import backend as K
from keras import regularizers, optimizers
from keras.layers import Dense, Dropout, Conv1D, \
AveragePooling1D, MaxPooling1D, LSTM, GRU, Flatten
from keras.callbacks import ReduceLROnPlateau

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, average_precision_score, \
roc_auc_score,confusion_matrix


```

```python
df1 = df[["course", "user_profile", "region", "course_level", "course_present_in_platform_in_days", "module_used_by_user_in_days", 'total_no_of_times_module_opened',
       'activities_engaged_wrt_modules', 'total_clicks',
       'assessments_completed', 'average_assessment_score', 'Hrabsence', 'Consultations', 'Grades', 'final_result']]
```

```python
df1.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
df1 = df1.apply(le.fit_transform)
```

```python
# Decision Tree
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics

X = df1.loc[:, df1.columns != 'final_result']
y = df1['final_result']
xTrain, xTest, yTrain, yTest = train_test_split(X, y,train_size = 0.75)

dt = tree.DecisionTreeClassifier(criterion='gini')
dt = dt.fit(xTrain, yTrain)
train_pred = dt.predict(xTrain)
test_pred = dt.predict(xTest)
print("Accuracy:{0:.3f}".format(metrics.accuracy_score(yTest, test_pred)),"\n")
```

```python
# Gradient Boosting Regression
from sklearn.ensemble import GradientBoostingRegressor

X = df1.loc[:, df1.columns != 'final_result']
y = df1['final_result']
xTrain, xTest, yTrain, yTest = train_test_split(X, y,train_size = 0.75)

gb = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=10, random_state=0, loss='huber')
gb = gb.fit(xTrain, yTrain)
print("Accuracy:{0:.3f}".format(gb.score(xTest, yTest)))
```

```python
# Random Forest
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

X = df1.loc[:, df1.columns != 'final_result']
y = df1['final_result']
xTrain, xTest, yTrain, yTest = train_test_split(X, y,train_size = 0.75)

rf = RandomForestClassifier(n_estimators=10,random_state=33)
rf = rf.fit(xTrain, yTrain)
train_pred = rf.predict(xTrain)
test_pred = rf.predict(xTest)
print("Accuracy:{0:.3f}".format(metrics.accuracy_score(yTest, test_pred)),"\n")
```

```python
# Function to plot accuracy

from sklearn.model_selection import learning_curve
import numpy as np

def plot_accuracy(model):

    train_sizes, train_scores, test_scores = learning_curve(model, xTrain, yTrain, n_jobs=-1, train_sizes=np.linspace(.1, 1.0, 5), verbose=0)

    train_mean = np.mean(train_scores, axis=1)
    train_std = np.std(train_scores, axis=1)
    test_mean = np.mean(test_scores, axis=1)
    test_std = np.std(test_scores, axis=1)
    
    plt.figure()
    plt.gca().invert_yaxis()
    plt.grid()
    plt.ylim(0.0, 1.1)
    plt.title("Accuracy Plot")
    plt.xlabel("Testing")
    plt.ylabel("Accuracy %")

    plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, alpha=0.1)
        
    plt.plot(train_sizes, test_mean, 'bo-', color = "r", label="Test Score")
```

```python
plot_accuracy(rf)
```

```python
#Random Forest
import scikitplot as skplt
from sklearn.model_selection import cross_val_predict

y_pred = cross_val_predict(rf, xTest, yTest)
skplt.metrics.plot_confusion_matrix(yTest, y_pred, normalize=True)

plt.show()
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(yTest, y_pred)
```

```python
print(classification_report(yTest, y_pred))
```
