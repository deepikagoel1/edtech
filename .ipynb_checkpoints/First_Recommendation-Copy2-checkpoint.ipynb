```python
Mean = data.groupby(by=["Courses_Name", 'course_level', 'id_user'],as_index=False)['Course_Rating'].mean()
Rating_avg = pd.merge(data,Mean,on='id_user')
Rating_avg['avg_rating']=Rating_avg['Course_Rating_x']-Rating_avg['Course_Rating_y']
Mean.head()
```

```python
Mean.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Mean['tags'] = Mean['Courses_Name'] + Mean['course_level']  + Mean['id_user'].astype('str')
```

```python
Mean

```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(Mean['tags']).toarray()
```

```python
import nltk #for stemming process
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#defining the stemming function
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
Mean['tags'] = Mean['tags'].apply(stem) #applying stemming on the tags column
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(course):
    course_index = Mean[Mean['Courses_Name'] == course].index[0]
    distances = similarity[course_index]
    course_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])
    
    for i in course_list:
        Mean['Recommendation'] = Mean.iloc[i[0]].Courses_Name
#         print(new_df.iloc[i[0]].Courses_Name)
```

```python
recommend('Create Your First Python Program') 
```
